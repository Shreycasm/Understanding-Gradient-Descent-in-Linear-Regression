```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_openml
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = fetch_openml(name='boston')
values = data.data
columns = data.feature_names
target = data.target
df = pd.concat([pd.DataFrame(values, columns=columns), pd.Series(target, name="target")], axis=1)
```

we will try to understand the Gradient Descent for Linear Regression 

Using Gradient Descent we usually try to minimise the loss function to get the optimal values of slope and intercepts

Steps to use Gradient Descent are:
 1) Start with random initial values for the model parameters.

 2) Calculate the gradient of the cost function with respect to each model parameter. This gradient points in the direction of the steepest increase in the cost.

 3) Adjust the model parameters by moving in the opposite direction of the gradient. The magnitude of the update is determined by a learning rate, which is a hyperparameter.

 4) Steps 2 and 3 are repeated iteratively until convergence or a predefined number of iterations. The process aims to reach the minimum of the cost function, where the gradient is close to zero.

 - At first I'll get the values of m and b using the Sklearns Linear Regression Class.
 - Then will keep the value of m constant and try to get the optimal value of b 
 - then do the opposite by keeping the value of b constant and try to get the optimal value of m 
 - in the end will update the both and find the optimal value for both m and b

Here ill pass only 1 input variable and the target variable to get the visualizations

The formulas are:
- $ \ y=mx+b$ - equation of line
- $ \sum_{i=1}^{n} (y_i - (mx_i + b))^2 $ - loss Function
- $b_{\text{new}} = b_{\text{old}} - \eta \cdot \ slope$ - formula to find value of b
- here slope is calculated by partially differeanting $ \sum_{i=1}^{n} (y - mx - b)^2$ with respect to b

```python
ax = sns.scatterplot(x = df["LSTAT"], y = df["target"], color="black")
```

```python
X_train, X_test, y_train, y_test = train_test_split(df["LSTAT"] , target, test_size=0.2, random_state=0)
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X_train.values.reshape(-1,1), y_train)
pred = linear_regression.predict(X_test.values.reshape(-1,1))
print(f"Coefficient of given model is: {linear_regression.coef_}")
print(f"Inpercept of given model is: {linear_regression.intercept_}")
print(f"Mean Square Error: {mean_squared_error(y_test, pred)} ")
```

```python
ax1= sns.scatterplot(x = df["LSTAT"], y = df["target"], color="black")

plt.plot(X_train , linear_regression.predict(X_train.values.reshape(-1,1)), color="red")
```

This is the regression line by OLS method

```python
b = 50
m = -0.95648761
lr = 0.0005

epochs = 10

bvalues = []
for i in range(epochs):
    loss_slope = -2 * np.sum(target - m * df["LSTAT"].ravel() - b)
    b = b - (lr * loss_slope)
    bvalues.append(b)
    y_pred = m * df["LSTAT"] + b
    plt.plot(df["LSTAT"] , y_pred)

ax = sns.scatterplot(x = df["LSTAT"], y = df["target"], color="black")
```

Here we can see by keeping the value of m as constant and choosing the random value for b the regression line tends to move towards the optimal line with increasing epochs and finally comes closer towards the optimal line

here the point to note is that aswe change the value of n or learnig rate the lines shows different behaviour which is shown in below plot
